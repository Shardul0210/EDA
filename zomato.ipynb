```python
df_country=pd.read_excel('Country-code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
## country-code and country in Country code file and country code in zomato file. 
# merge them to get country in zomato 
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_value=final_df.Country.value_counts().values
country_value
```

```python
## Pie chart
plt.pie(country_value,labels=country_names)
```

```python
## Pie chart - Top 3 countries
explode=[0.1,0,0]
plt.pie(country_value[:3],labels=country_names[:3],explode=explode,autopct='%1.2f%%')
```

# Observation 
Zomato maximum records or transactions are from India amd United Kingdoms

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()  
```

```python
rating=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
rating
```

# Observations
1. when rating between 4.5 to 4.9 --->excellent
2.when rating between 4.0 to 4.4 --->very good
3.when rating between 3.5 to 3.9 --->good
4.when rating between 2.5 to 3.4 --->avg
5.when rating between 1.8 to 2.4 --->poor
6.when rating between 0.0 ---> not rated
